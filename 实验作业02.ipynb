```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 打开文件并读取数据
with open('计算机学院2022年硕士研究生复试成绩公示.txt', 'r',encoding="gbk") as f:
    data = f.readlines()
columns=['考生编号', '姓名', '报考学院', '报考专业', '报考学习方式', '专项计划', '初试总分', '复试总分', '综合成绩', '备注']#列标题
# 划分每11个学生信息为一行
result=[]
for i in range(0, len(data)):
    data[i]=data[i].replace('\n','')
    line=data[i].split(" ")
    del line[:11]
    for i in range(0, len(line), 10):
        row = line[i:i+10]  # 切分得到一行数据
        row.append("")
        result.append(row)   # 将该行数据添加到二维列表中
result=np.array(result)
df= pd.DataFrame(result[:,1:], columns=columns)
df['序号']=result[:,0]
df.set_index('序号',inplace=True)
df
```

```python
#新增一列，初始化复试结果为“待定”
df['复试结果']="待定"

#计算录取的人数
xue=59-46
zhuan=99-29
# total=xue+zhuan
```

```python
df.loc[:,'综合成绩'] = pd.to_numeric(df.loc[:, '综合成绩'])  # 确保综合成绩列是数字类型
# 选取报考专业为'081200计算机科学与技术'的所有行，并按照综合成绩的降序排列
top_xue = df[df['报考专业'] == '081200计算机科学与技术'].sort_values(by='综合成绩',ascending=False).iloc[:xue]  # 选出综合成绩排名前13的学硕学生
df.loc[top_xue.index, '复试结果'] = '录取'  # 修改这13个学生的复试结果为“录取”
top_zhuan = df[df['报考专业'] == '085400电子信息'].sort_values(by='综合成绩',ascending=False).iloc[:zhuan]  # 选出综合成绩排名前70的专硕学生
df.loc[top_zhuan.index, '复试结果'] = '录取'  
```

```python
#将符合特殊情况下的学生修改为录取
special=['少高','退役士兵']
sp_con=df[(df['专项计划'].isin(special))&(df['复试总分'].astype(float) > 60)]
df.loc[sp_con.index,'复试结果']='录取'
df
```

```python
#筛选出可以参与奖学金评比的学生
eligible_students = df[(df['专项计划'].isin(["无"])) & (df['复试结果'] == '录取')]
eligible_students
```

```python
#已知学硕推免有46，奖学金比例分别为A：0.6，B：0.2，C：0.2
#首先计算出考研能获得A等奖学金的人数:int(59*0.6)-46
#由于int(59*0.6)-46<0，因此学硕考研录取的同学只能平分B等和C等奖学金

cs_quota_B_last=int(59*0.2)
cs_stu=eligible_students[eligible_students['报考专业'] == '081200计算机科学与技术'].sort_values(by='综合成绩',ascending=False)
cs_B=cs_stu.iloc[:cs_quota_B_last]
df.loc[cs_B.index,'奖学金等级']='B等'
cs_C=cs_stu.iloc[cs_quota_B_last:]
df.loc[cs_C.index,'奖学金等级']='C等'

#已知专硕推免有29，奖学金比例分别为A：0.6，B：0.2，C：0.2
#首先计算出考研能获得A等奖学金的人数:int(59*0.6)-46
#由于int(59*0.6)-46<0，因此学硕考研录取的同学只能平分B等和C等奖学金
ee_quota_A_last=int(99*0.6)-29
ee_stu=eligible_students[eligible_students['报考专业'] == '085400电子信息'].sort_values(by='综合成绩',ascending=False)
ee_A=ee_stu.iloc[:ee_quota_A_last]
df.loc[ee_A.index,'奖学金等级']='A等'
ee_quota_B_last=int(99*0.2)
ee_B=ee_stu.iloc[ee_quota_A_last:ee_quota_A_last+ee_quota_B_last]
df.loc[ee_B.index,'奖学金等级']='B等'
ee_C=ee_stu.iloc[ee_quota_A_last+ee_quota_B_last:]
df.loc[ee_C.index,'奖学金等级']='C等'

df[df['报考专业']=='081200计算机科学与技术']
```
